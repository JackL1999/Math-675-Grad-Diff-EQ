```python
import matplotlib.pyplot as plt
import numpy as np

# Define the system of equations
def system(X, Y):
    return X, Y  # x' = x, y' = y

# Points requested for change vectors
points = [(1, 1), (1, -1), (-1, -1), (-1, 1)]

# Create a grid for the vector field
x = np.linspace(-2, 2, 15)
y = np.linspace(-2, 2, 15)
X, Y = np.meshgrid(x, y)
U, V = system(X, Y)

plt.figure(figsize=(8, 8))

# Plot the general vector field (streamlines/arrows)
plt.quiver(X, Y, U, V, color='lightgray', alpha=0.5)

# Plot the specific change vectors requested
for px, py in points:
    dx, dy = system(px, py)
    plt.quiver(px, py, dx, dy, color='red', angles='xy', scale_units='xy', scale=1)
    plt.plot(px, py, 'bo') # mark the state point
    plt.text(px, py, f' ({px},{py})', verticalalignment='bottom')

plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.title("State Space Change Vectors for x'=x, y'=y")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, linestyle='--')
plt.xlim([-2.5, 2.5])
plt.ylim([-2.5, 2.5])
plt.show()
```

## Question 1: Graph a

```python
import matplotlib.pyplot as plt
import numpy as np

a, b = -1, 0, 
c, d = 0, -5

points = [(1, 1), (1, -1), (-1, -1), (-1, 1)]
x_vals, y_vals = map(np.array, zip(*points))

x_prime = a * x_vals + b * y_vals
y_prime = c * x_vals + d * y_vals


plt.figure(figsize=(6, 6))
plt.scatter(x_vals, y_vals, color='red', zorder=5)
plt.quiver(x_vals, y_vals, x_prime, y_prime, 
           angles='xy', scale_units='xy', scale=1, color='blue')


styles = [
    ('left', 'bottom', (10, 5)),   
    ('left', 'top', (10, -5)),     
    ('right', 'top', (-10, -5)),   
    ('right', 'bottom', (-10, 5))  
]

for (x, y), (ha, va, offset) in zip(points, styles):
    plt.annotate(f"({x}, {y})", 
                 xy=(x, y), 
                 xytext=offset, 
                 textcoords='offset points',
                 ha=ha, va=va,
                 color='red', fontweight='bold', fontsize=12)

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid(True, linestyle='--', alpha=0.7)
plt.savefig("HW_1_Q1a.pdf", bbox_inches='tight')
plt.show()
```
