```python
import matplotlib.pyplot as plt
import numpy as np

# Define the system of equations
def system(X, Y):
    return X, Y  # x' = x, y' = y

# Points requested for change vectors
points = [(1, 1), (1, -1), (-1, -1), (-1, 1)]

# Create a grid for the vector field
x = np.linspace(-2, 2, 15)
y = np.linspace(-2, 2, 15)
X, Y = np.meshgrid(x, y)
U, V = system(X, Y)

plt.figure(figsize=(8, 8))

# Plot the general vector field (streamlines/arrows)
plt.quiver(X, Y, U, V, color='lightgray', alpha=0.5)

# Plot the specific change vectors requested
for px, py in points:
    dx, dy = system(px, py)
    plt.quiver(px, py, dx, dy, color='red', angles='xy', scale_units='xy', scale=1)
    plt.plot(px, py, 'bo') # mark the state point
    plt.text(px, py, f' ({px},{py})', verticalalignment='bottom')

plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.title("State Space Change Vectors for x'=x, y'=y")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True, linestyle='--')
plt.xlim([-2.5, 2.5])
plt.ylim([-2.5, 2.5])
plt.show()
```

```python
import matplotlib.pyplot as plt

a = 2
b = 0
c = 0
d = 1

x_values = np.array([1, 1, -1, -1])
y_values = np.array([1, -1, -1, 1])

x_prime_values = a*x_values + b*y_values
y_prime_values = c*x_values + d*y_values

plt.quiver(x_values, y_values, x_prime_values, y_prime_values, 
           angles='xy', scale_units='xy', scale = 1, color='blue')

plt.axhline(0, color='black')
plt.axvline(0, color='black')
plt.xlim(-7, 7)
plt.ylim(-5, 5)
plt.grid(True)
plt.savefig('change_vectors.png')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# System parameters (Matrix elements)
# Corresponding to: 
# dx/dt = ax + by
# dy/dt = cx + dy
a, b = 2, 0
c, d = 0, 1

# 1. Create a grid of points
x = np.linspace(-5, 5, 20)
y = np.linspace(-5, 5, 20)
X, Y = np.meshgrid(x, y)

# 2. Calculate the derivatives (the "flow" at each point)
U = a*X + b*Y
V = c*X + d*Y

# 3. Optional: Normalize the vectors 
# This makes all arrows the same length so you can see the direction 
# clearly even where the growth is very fast or slow.
N = np.sqrt(U**2 + V**2)
U, V = U/N, V/N

# 4. Plotting
plt.figure(figsize=(8, 6))
plt.quiver(X, Y, U, V, color='blue', alpha=0.8, pivot='mid')

plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.title(f'Phase Portrait: dx/dt={a}x, dy/dt={d}y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.6)
plt.savefig('phase_portrait.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# System parameters (Matrix elements)
a, b = 2, 0
c, d = 0, 1

# 1. Create a grid of points
# For stream plots, a denser 1D array for x and y works best
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# 2. Calculate the derivatives
U = a*X + b*Y
V = c*X + d*Y

# 3. Create the stream plot
plt.figure(figsize=(6, 6))

# streamplot takes the 1D arrays x, y and the 2D arrays U, V
plt.streamplot(x, y, U, V, color='blue', linewidth=1, density=1, arrowstyle='->', arrowsize=1.5)

# 4. Styling
plt.axhline(0, color='black', lw=1.5)
plt.axvline(0, color='black', lw=1.5)
plt.title(f'Phase Portrait (Streamlines): dx/dt={a}x, dy/dt={d}y')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(-5, 5)
plt.ylim(-5, 5)

# Save the output
plt.savefig('phase_portrait_streamlines.png')
```
